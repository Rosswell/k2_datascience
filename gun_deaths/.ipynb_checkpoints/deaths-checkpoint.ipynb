### Assignment

- Import the csv
- Read it into a list
- Preview the first 5 entries

```python
gunslist = []
with open("guns.csv") as filename:
    for line in filename:
        line = line.strip().split(',')
        gunslist.append([x[1:-1] if x.startswith('\"') else x for x in line])
for x in range(5):
    print(gunslist[x])
```
